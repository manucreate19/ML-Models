```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', None],
    'Age': [25, 30, None, 45, 28, 22],
    'Gender': ['F', 'M', 'M', 'M', 'F', 'F'],
    'Salary': [50000, 60000, 58000, None, 52000, 51000]
}

df = pd.DataFrame(data)
df
```

```python
df.isnull().sum()
df.dropna()
```

```python
df.fillna({
    'Name': 'Unknown',
    'Age': df['Age'].mean(),
    'Salary': df['Salary'].median()
}, inplace=True)

df
```

```python
new_row = pd.DataFrame([{'Name': 'Alice', 'Age': 25, 'Gender': 'F', 'Salary': 50000}])

# Append to existing df
df = pd.concat([df, new_row], ignore_index=True)
print(df.duplicated())
print(df)

# Remove duplicates
df = df.drop_duplicates()

# Show cleaned DataFrame
print(df)
```

```python
df.rename(columns={'Salary': 'Monthly Salary'}, inplace=True)
df['Age'] = df['Age'].astype(float)
df['Yearly Salary'] = df['Monthly Salary'] * 12
df
```
